```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

**Matplotlib Basics**

```python
x = np.linspace(0, 5, 11)
x
```

```python
y = x ** 2
y
```

```python
plt.plot(x, y)
plt.title("Hello")
plt.xlabel(" X axis")
plt.ylabel(" Y axis")
```

**Subplots in Matplotlib**

```python
plt.subplot(2, 2, 1)
plt.plot(x, y)
plt.subplot(2, 2, 2)
plt.plot(y, x)
plt.subplot(2, 2, 3)
plt.plot(x, x**3)
plt.subplot(2, 2, 4)
plt.plot(x, x**0.5)
```
