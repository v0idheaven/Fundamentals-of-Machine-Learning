```python
# what is the capital city of the country with the largest population?
df[df['population'] == df['population'].max()]['capital_city']
```

```python
# Which country has the least population?
df[df['population'] == df['population'].min()]['country']
```

```python
#what is the capital of the country with the least population?
df[df['population'] == df['population'].min()]['capital_city']
```

```python
# give me top 5 country with highest democtaric score
df.sort_values(by = 'democracy_score', ascending=False, inplace=True)
```

```python
# How many total regions are there
df['region'].value_counts().count()
```

```python
# How many countries lies in eastern europe region?
df[df['region'] == 'Eastern Europe']['country']
```

```python
# Who is the leader of the 2nd highest populated country?
df[df['population'] == df['population'].nlargest(2).iloc[1]]['political_leader']
```

```python
# how many countries are there whose political leader are unknown?
df[df['political_leader'].isna()]['country'].count()
```

```python
# How many country have Republic in their name?
df[df['country_long'].str.contains('Republic')]['country'].count()
```

```python
# How many country have Republic in their name?
count = 0
def counting(txt):
    global count
    if 'republic' in txt.lower():
        count += 1
    return txt

df['country_long'].apply(counting)
print(count)
```

```python
# Which country in african region has highest population?
africa_df = df[df['continent'] == 'Africa']
```

```python
africa_df[africa_df['population'] == africa_df['population'].max()]['country']
```

```python
# how many countries have a democracy score of 8 or higher?
df[df['democracy_score'] >= 8]['country'].count()
```
